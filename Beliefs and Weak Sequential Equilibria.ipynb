# Behavioral Strategies, Beliefs, and Sequential Equilibrium
Recall, a behavioral strategy is a probability distribution over actions at every information set:
* Player 1: $(p(L),p(R))$
* Player 2: $[p(A),p(B)|(p(E),p(F)]$
* Player 3: $[p(C),p(D)|(p(G),p(H)]$

![Sequential Game](img/seqgame1.png)

**Strategy profile:** a behavioral strategy for each player

*Kuhn's Theorem:* there is a behavioral strategy that is equivalent to a mixed strategy

## Subgame Perfect Nash Equilibria, Revisited
**Every finite extensive form game has at least one subgame-perfect equilibrium in mixed strategies**

* In first proper subgame (following L): 
    - Player 2 BR: (p(A),p(B)) = (0.5,0.5)
    - Player 3 BR: (p(C),p(D))= (0.25,0.75)
* In second proper subgame (following R): 
    - Player 2 BR: (p(E),p(F)) = (0.67,0.33)
    - Player 3 BR: (p(G),p(H))= (0.33,0.67)
* Given BR by 2 and 3, player 1‚Äôs BR is to pick L (higher exp payoff)

## Beliefs and Assessment 
An assessment of an extensive form game is a pair $(\sigma,\mu)$ where $\sigma$ is a profile of behavioral strategies and $\mu$ is a belief system:
* $\mu$ = a list of probability distributions, one for every information set, over the likelihood of the nodes in the information set

### Example, cont. 
* Player 1: Information set: I={‚¶∞} 
    - Beliefs p(‚¶∞)=1
* Player 2, Information set 1: I={L}
    - Beliefs p(L)=1
* Player 2, Information set 2: I ={R}
    - p(R) =1
* Player 3, Information set 3: I={A,B}
    - Beliefs (p(A),p(B))
* Player 3, Information set 3: I={E,F}
    - Beliefs (p(E),p(F))
* An assessment for this game could be:
$$\sigma=[(1,0),(0.5,0.5|1,0),(1,0|0,1)]$$
$$\mu = [(1),(1|1),(0.5,0.5|1,0)] $$

## Sequentially Rational Behavioral Strategies
Given an extensive form game with assessment $(\sigma,\mu)$, we say that player i‚Äôs behavioral strategy $\sigma_i$  is sequentially rational if the expected payoff given the beliefs of the player is at least as big as any other behavioral strategy available to player $i$

### Example, Cont
Given the assessment above:
1. Player 3: she believes that there is a 50-50 chance of A occurring, and she believes that there is a 100% chance of E occurring
$$E(C) = 0.5*1+0.5*3= 2$$
$$E(D) = 2 $$

$$E(G) = 3$$
$$E(H) = 2$$
The choice $[p(C),p(D)|p(G),p(H)]=[(p,1-p)|(1,0)]$ is sequentially rational. 

For player 2, she believes that there is a 100% chance of L occurring. 
Selecting p(A),p(B) = (0.5,0.5) implies that the expected value of A and B are equal. 
Since player 3 is mixing: 
$$E(A) =(0.5*3+0.5*0)=1.5$$
$$E(B) =(0.5*0+0.5*1)=0.5$$
(¬Ω,¬Ω|1,0) is not sequentially rational

## Bayesian Updating
Beliefs are considered ‚Äúconsistent‚Äù if they are fully updated using Bayes‚Äô Rule.  (for information sets that are reached)
Let x and y be nodes in a player‚Äôs information set. Then
$$p(h|I_i) = \frac{p(h)}{\sum_{h\in I_i} p(h)}$$

Which is to say that the belief you have that $h$ occurred must be equal to the fraction of times that the previous player picks the history. 
#### Example: If player 2 picks A 1/2 of the time and B 1/2 of the time, then the belief that is consistent with Bayes' rule is: 
$$p(A|\{A,B\}) = \frac{0.5}{0.5+0.5}=0.5$$
Thus the beliefs of player 2 are fully updated.

### Example, Cont
Let‚Äôs update all the beliefs here:
$$\sigma =[(1,0),(¬Ω,¬Ω|1,0),(1,0|0,1)]$$ 
$$\mu = [(1),(1|1),(0.5,0.5|1,0)] $$
If ùõî1=(1,0), then the belief ùõç2 = (1|1) is updated because the information set reached is info set 1  
If ùõî2=(¬Ω,¬Ω|1,0), then belief ùõç3 =(¬Ω,¬Ω|1,0) is fully updated 

## Weak Sequential Equilibrium 
A **weak sequential equilibrium** is an assessment (ùõî,ùõç) which satisfies sequential rationality and Bayesian updating at all information sets

### New Example
![Sequential Game](img/seqgame2.png)
**Case 1**: Player 1 picks  ùõî1 = (1,0|1,0) 
* Player 2s beliefs are ùõç2 = (0,1) since C is only way to get to info set
    - Given beliefs, player 2 picks ùõî2 = (1,0) 
    - But if player 2 picks E, then A is no longer sequentially rational. 
* There is no WSE with ùõî1 = (1,0|1,0)  

**Case 2**: Player 1 picks  ùõî1 = (0,1|0,1) 
* Player 2s beliefs are ùõç2 = (1,0) since B is only way to get to info set
    - Given beliefs, player 2 picks ùõî2 = (0,1) 
    - But if player 2 picks F, then B is no longer sequentially rational. 
* There is no WSE with ùõî1 = (0,1|0,1)  

**Case 3**: Player 1 picks  ùõî1 = (0,1|1,0)

$$E(B) = 2p + 0(1-p) \geq 1=E(A)$$
$$p\geq 0.5$$

$$ E(C) = 3p + 0(1-p) \geq 2=E(D)$$ 
$$ p\geq \frac{2}{3}$$  
* Weak sequential equilibrium: 
$$ ùõî = [(0,1),(1,0),(p>= \frac{2}{3})] \qquad ùõç=( \frac{1}{4}, \frac{3}{4}) $$

**Case 4**: Player 1 picks  ùõî1 = (1,0|0,1) 
* Player 2s beliefs are ùõç2 = (q,1-q ) 
* Player 2 ùõî2 = (p,1-p)

$$E(F) = q(3)+(1-q)(0) > 0q + 1(1-q) =E(E)$$
$$q\geq \frac{1}{4}$$ 
* Weak sequential equilibrium: 
$$ ùõî = [(1,0),(0,1),(0,1)] \qquad ùõç=(q\geq \frac{1}{4})$$
